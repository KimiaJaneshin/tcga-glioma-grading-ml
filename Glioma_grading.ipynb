```python
# =========================================================
# INSTALLS
# =========================================================
!pip install -q kagglehub xgboost lightgbm shap pytorch-tabnet

import kagglehub
import numpy as np
import pandas as pd
import shap
import matplotlib.pyplot as plt
import seaborn as sns
import torch
import os

from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import (
    roc_auc_score, accuracy_score, f1_score,
    precision_score, recall_score, brier_score_loss,
    confusion_matrix, precision_recall_curve
)

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
import lightgbm as lgb
from pytorch_tabnet.tab_model import TabNetClassifier

RANDOM_STATE = 42
N_SPLITS = 5

SCALED_MODELS = {"Logistic Regression", "TabNet"}
# =========================================================
# DOWNLOAD + LOAD DATA
# =========================================================
path = kagglehub.dataset_download("vinayjose/glioma-grading-clinical-and-mutation-features")

print("Path to dataset files:", path)

df = pd.read_csv(os.path.join(path, "TCGA_InfoWithGrade.csv"))

TARGET = "Grade"

print("Original Grade values:")
print(df[TARGET].value_counts(dropna=False))
print("Unique values:", df[TARGET].unique())

# If Grade is already numeric, don't map it
if pd.api.types.is_numeric_dtype(df[TARGET]):
    print("Grade is already numeric.")
else:
    df[TARGET] = df[TARGET].astype(str).str.strip().str.upper()

    print("\nCleaned Grade values:")
    print(df[TARGET].value_counts(dropna=False))

    # only map if labels are truly LGG / GBM
    if set(df[TARGET].dropna().unique()).issubset({"LGG", "GBM"}):
        df[TARGET] = df[TARGET].map({"LGG": 0, "GBM": 1})
    else:
        print("\nGrade values are not LGG/GBM, so custom mapping is needed.")
        print("Current unique values:", df[TARGET].dropna().unique())
print("\nAfter possible mapping:")
print(df[TARGET].value_counts(dropna=False))

df = df.dropna(subset=[TARGET]).copy()

print("\nShape after dropping missing target:", df.shape)

X = df.drop(columns=[TARGET])
X = pd.get_dummies(X, drop_first=True)
X = X.fillna(0)
y = df[TARGET].values

print("X shape:", X.shape)
print("y shape:", y.shape)

# ==============================
# CROSS VALIDATION
# ==============================

skf = StratifiedKFold(
    n_splits=N_SPLITS,
    shuffle=True,
    random_state=RANDOM_STATE
)

models = {

    "Logistic Regression":
        LogisticRegression(max_iter=2000),

    "Random Forest":
        RandomForestClassifier(
            n_estimators=800,
            max_depth=10,
            class_weight="balanced",
            random_state=RANDOM_STATE
        ),

    "XGBoost":
        XGBClassifier(
            n_estimators=500,
            learning_rate=0.03,
            max_depth=4,
            subsample=0.9,
            colsample_bytree=0.9,
            reg_lambda=1.0,
            reg_alpha=0.2,
            eval_metric="logloss",
            random_state=RANDOM_STATE
        ),

    "LightGBM":
        lgb.LGBMClassifier(
            n_estimators=300,
            learning_rate=0.03,
            num_leaves=15,
            max_depth=4,
            subsample=0.9,
            colsample_bytree=0.9,
            reg_alpha=0.2,
            reg_lambda=1.0,
            verbose=-1,
            random_state=RANDOM_STATE
        ),

    "TabNet":
        TabNetClassifier(
            n_d=64,
            n_a=64,
            n_steps=5,
            gamma=1.7,
            lambda_sparse=1e-4,
            optimizer_fn=torch.optim.Adam,
            optimizer_params=dict(lr=2e-3),
            seed=RANDOM_STATE
        )
}

oof_preds = {name: np.zeros(len(y)) for name in models}
results = []

# ==============================
# TRAINING
# ==============================
# ==============================
# TRAINING (clean scaling)
# ==============================

for name, model in models.items():
    print("\nTraining", name)
    fold_aucs = []

    for fold, (train_idx, test_idx) in enumerate(skf.split(X, y), start=1):
        X_train_raw, X_test_raw = X.iloc[train_idx], X.iloc[test_idx]
        y_train, y_test = y[train_idx], y[test_idx]

        # Scale only when needed
        if name in SCALED_MODELS:
            scaler = StandardScaler()
            X_train = scaler.fit_transform(X_train_raw)
            X_test  = scaler.transform(X_test_raw)
        else:
            # keep as numpy arrays for consistent downstream handling
            X_train = X_train_raw.to_numpy()
            X_test  = X_test_raw.to_numpy()

        if name == "TabNet":
            # TabNet should be re-created each fold (fresh weights)
            tabnet = TabNetClassifier(
                n_d=64,
                n_a=64,
                n_steps=5,
                gamma=1.7,
                lambda_sparse=1e-4,
                optimizer_fn=torch.optim.Adam,
                optimizer_params=dict(lr=2e-3),
                seed=RANDOM_STATE
            )

            tabnet.fit(
                X_train,
                y_train,
                eval_set=[(X_test, y_test)],
                eval_metric=["auc"],
                patience=20,
                max_epochs=200,
                batch_size=256
            )
            preds = tabnet.predict_proba(X_test)[:, 1]

        else:
            # For sklearn / xgboost / lgbm
            model.fit(X_train, y_train)
            preds = model.predict_proba(X_test)[:, 1]

        oof_preds[name][test_idx] = preds

        auc = roc_auc_score(y_test, preds)
        fold_aucs.append(auc)
        print(f"Fold {fold} AUC: {auc:.4f}")

    print(f"{name} Mean AUC: {np.mean(fold_aucs):.4f}")


# ==============================
# THRESHOLD OPTIMIZATION
# ==============================

best_thresholds = {}

for name, probs in oof_preds.items():

    precision, recall, thresholds = precision_recall_curve(y, probs)

    f1_scores = 2 * precision * recall / (precision + recall + 1e-9)

    best_idx = np.argmax(f1_scores)

    best_thresholds[name] = thresholds[best_idx]

    print(
        name,
        "Best threshold:",
        round(thresholds[best_idx],4),
        "Best F1:",
        round(f1_scores[best_idx],4)
    )

# ==============================
# FINAL METRICS
# ==============================

for name, probs in oof_preds.items():

    threshold = best_thresholds[name]

    preds = (probs >= threshold).astype(int)

    auc = roc_auc_score(y, probs)
    acc = accuracy_score(y, preds)
    f1 = f1_score(y, preds)
    precision = precision_score(y, preds)
    recall = recall_score(y, preds)
    brier = brier_score_loss(y, probs)

    results.append({
        "Model": name,
        "AUC": auc,
        "Accuracy": acc,
        "F1": f1,
        "Precision": precision,
        "Recall": recall,
        "Brier": brier
    })

results_df = pd.DataFrame(results).sort_values("AUC", ascending=False)

print("\nFinal Results")
print(results_df)

# ==============================
# CONFUSION MATRICES
# ==============================

for name, probs in oof_preds.items():
    preds = (probs >= best_thresholds[name]).astype(int)
    cm = confusion_matrix(y, preds)

    plt.figure(figsize=(6, 5))
    sns.heatmap(
        cm,
        annot=True,
        fmt="d",
        cmap="Blues",
        cbar=False,
        xticklabels=["LGG (0)", "GBM (1)"],
        yticklabels=["LGG (0)", "GBM (1)"],
        annot_kws={"size": 16, "weight": "bold"}
    )

    plt.title(f"{name}\nConfusion Matrix", fontsize=14, pad=12)
    plt.xlabel("Predicted Grade", fontsize=12, labelpad=10)
    plt.ylabel("True Grade", fontsize=12, labelpad=10)


    plt.tight_layout()

    filename = name.lower().replace(" ", "_") + "_cm_clean.png"
    plt.savefig(filename, dpi=300, bbox_inches="tight")
    plt.show()

# ==============================
# BOOTSTRAP AUC CONFIDENCE INTERVAL
# ==============================

def bootstrap_auc(y_true, y_pred, n_bootstrap=2000):

    rng = np.random.RandomState(42)
    scores = []

    for i in range(n_bootstrap):

        idx = rng.randint(0, len(y_true), len(y_true))

        if len(np.unique(y_true[idx])) < 2:
            continue

        score = roc_auc_score(y_true[idx], y_pred[idx])
        scores.append(score)

    return np.percentile(scores, [2.5, 97.5])

print("\nAUC 95% Confidence Intervals")

for name, probs in oof_preds.items():

    ci = bootstrap_auc(y, probs)

    print(name, "CI:", ci)

# ==============================
# DECISION CURVE ANALYSIS (Two-Panel Portfolio Version)
# ==============================

import numpy as np
import matplotlib.pyplot as plt

thresholds = np.linspace(0.01, 0.99, 100)
n = len(y)
prevalence = np.mean(y)


curves = {}
for name, probs in oof_preds.items():
    net_benefits = []
    for t in thresholds:
        preds = (probs >= t).astype(int)
        tp = np.sum((preds == 1) & (y == 1))
        fp = np.sum((preds == 1) & (y == 0))
        nb = (tp / n) - (fp / n) * (t / (1 - t))
        net_benefits.append(nb)
    curves[name] = np.array(net_benefits)

treat_all = prevalence - (1 - prevalence) * (thresholds / (1 - thresholds))
treat_none = np.zeros_like(thresholds)


all_model_vals = np.concatenate(list(curves.values()))
ymax_full = np.max(all_model_vals) + 0.05
ymin_full = -0.10

# --- Start Plotting ---
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(11, 14))

# PLOT 1: FULL OVERVIEW (Panel A)
for name, nb in curves.items():
    ax1.plot(thresholds, nb, label=name, lw=2.5)

ax1.plot(thresholds, treat_none, color="black", linestyle="--", lw=1.5, label="Treat None")
ax1.plot(thresholds, treat_all, color="gray", linestyle="--", lw=1.5, alpha=0.7, label="Treat All")

ax1.set_xlim(0.0, 1.0)
ax1.set_ylim(ymin_full, ymax_full)
ax1.set_title("Decision Curve Analysis: Full Perspective (0.0 - 1.0)", fontsize=15, fontweight='bold', pad=15)
ax1.set_ylabel("Net Benefit", fontsize=12)
ax1.legend(loc="upper right", frameon=True, fontsize=10)
ax1.grid(True, alpha=0.3)

# PLOT 2: CLINICAL ZOOM (Panel B)
for name, nb in curves.items():
    ax2.plot(thresholds, nb, label=name, lw=2.5)

ax2.plot(thresholds, treat_none, color="black", linestyle="--", lw=1.5, label="Treat None")
ax2.plot(thresholds, treat_all, color="gray", linestyle="--", lw=1.5, alpha=0.7, label="Treat All")

ax2.set_xlim(0.50, 0.85)
ax2.set_ylim(-0.02, 0.20)
ax2.set_title("Clinical Utility: Target Threshold Zoom (0.50 - 0.85)", fontsize=15, fontweight='bold', pad=15)
ax2.set_xlabel("Threshold Probability", fontsize=12)
ax2.set_ylabel("Net Benefit", fontsize=12)
ax2.legend(loc="lower left", frameon=True, fontsize=10)
ax2.grid(True, alpha=0.3)

plt.tight_layout(pad=4.0)
plt.savefig("dca_portfolio_dual_panel.png", dpi=300, bbox_inches="tight")
plt.show()


# ==============================
# SHAP ANALYSIS (XGBoost)
# ==============================

print("\nRunning SHAP analysis")

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

xgb_model = XGBClassifier(
    n_estimators=500,
    learning_rate=0.03,
    max_depth=4,
    subsample=0.9,
    colsample_bytree=0.9,
    reg_lambda=1.0,
    reg_alpha=0.2,
    eval_metric="logloss",
    random_state=42
)

xgb_model.fit(X_scaled, y)

explainer = shap.TreeExplainer(xgb_model)

shap_values = explainer.shap_values(X_scaled)

shap.summary_plot(shap_values, X, show=False)
plt.title("SHAP Feature Importance (XGBoost)", fontsize=14, pad=12)
plt.tight_layout()
plt.savefig("shap_xgboost_summary.png", dpi=300, bbox_inches="tight")
plt.show()

```
